--------------------------
- iloc의 여러 개 컬럼: df.iloc[:, df[]]
- loc의 여러 개 컬럼: df.loc[:, []]

```python
#1번
movies.loc[:, ['director_name', 'gross', 'budget' ]]

#2번
movies.loc[movies['duration'] >= 170 ]

#3번
movies.loc[movies['director_facebook_likes'] == 0 ]

#4번
movies.loc[(movies['imdb_score'] >= 8.0) & (movies['country'] == 'USA')]

#5번
movies.loc[:, 'num_critic_for_reviews'].mean()
```

1. 'actor_1_name', 'actor_2_name', 'actor_3_name' 열의 데이터 중에서 'Tom Hanks'가 포함된 모든 데이터를 선택하는 코드를 작성해보세요.
2. 'gross', 'budget' 열의 값을 더한 'total_profit' 열을 추가하는 코드를 작성해보세요.
3. 'country' 열의 값이 'USA'가 아닌 데이터를 모두 선택하고, 그 중에서 'imdb_score' 값이 7.0 이상인 데이터를 선택하는 코드를 작성해보세요.
4. 'director_name' 열의 값이 'Christopher Nolan'인 데이터를 선택하고, 그 중에서 'imdb_score' 값이 가장 높은 데이터를 선택하는 코드를 작성해보세요.
5. 'title_year' 열의 값이 2000년부터 2010년 사이인 데이터를 선택하고, 그 중에서 'imdb_score' 값이 가장 높은 상위 5개의 데이터를 선택하는 코드를 작성해보세요.

```python

#1번
movies.loc[movies[['actor_1_name', 'actor_2_name', 'actor_3_name']].isin(['Tom Hanks']).any(axis=1)]

#2번
movies['total_profit'] = movies['gross'] + movies['budget']
movies['total_profit']

#3번
movies.loc[(movies['country'] != 'USA') & (movies['imdb_score'] >= 7.0)]

#4번
movies.groupby('Christopher Nolan')['imdb_score']

#5번
year_movies = movies.loc[(movies['title_year'] >= 2000) & (movies['title_year'] <= 2010)]
year_movies.nlargest(5, 'imdb_score')
```

#filter함수 연습

```python
import pandas as pd

data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
        'Age': [25, 32, 18, 47, 28],
        'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'San Francisco'],
        'Salary': [75000, 95000, 45000, 120000, 65000],
        'Gender': ['F', 'M', 'M', 'M', 'F']}

df = pd.DataFrame(data)
'''
이 데이터프레임에서 나이(Age)가 30세보다 작고, 
성별(Gender)이 남성(M)인 데이터만 추출해보세요. 
추출한 데이터프레임의 이름은 result로 해주세요. 
단, filter 메소드를 사용해서 해결해보세요.
'''
df.head()
```

```python
result = df.filter(items=['Age', 'Gender']).query('Age < 30 & Gender == "M"')
result
```

```python
#1번
#이 데이터프레임에서 성별(Gender)이 여성(F)이고, 연봉(Salary)이 70000 이상인 데이터만 추출해보세요. 
#추출한 데이터프레임의 이름은 result1로 해주세요. 
#단, filter 메소드를 사용해서 해결해보세요.

result1 = df.filter(items = ['Gender', 'Salary']).query('Gender == "F" & Salary >= 70000')
result1
```

```python
#2번
#'Name'과 'City' 컬럼만 추출해보세요.
#추출한 데이터프레임의 이름은 result1로 해주세요.

#result2 = df.loc[:, ['Name', 'City']]
#result2

#같은 결과
result2 = df.filter(items=['Name', 'City'])
result2
```

```python
#3번
#'Name' 컬럼에서 'a'가 들어간 이름만 추출해보세요. 
#추출한 데이터프레임의 이름은 result2로 해주세요.

result3 = df.filter(items=['Name']).query('Name == "a"')
result3

#이건 시팔 왜 안되는건데
```

```python
#4번
#'Age'가 30 이상이고, 'Salary'가 80000 이상인 데이터만 추출해보세요. 
#추출한 데이터프레임의 이름은 result3으로 해주세요.

result4 = df.filter(items = ['Age', 'Salary']).query('Age >= 30 & Salary >= 80000')
result4
```
